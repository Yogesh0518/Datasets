```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Use k-means clustering with k=3 (since there are three classes in the Iris dataset)
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

# Reduce the dimensionality for visualization (using PCA)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualize the clusters
plt.figure(figsize=(10, 6))

# Plot the original data points
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y,label='True Labels')

# Plot the cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', label='Cluster Centers')

# Plot the cluster-assigned data points
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, label='Cluster Assignments')

plt.title('K-Means Clustering on Iris Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```
