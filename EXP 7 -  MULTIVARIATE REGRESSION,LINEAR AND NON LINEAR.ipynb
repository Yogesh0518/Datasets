```python
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Leechita G\Downloads\multiple_linear_regression_dataset.csv")
df
```

```python
X=df.drop(columns=['income'],axis=1)
y=df['income']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
```

```python
pred=lm.predict(X_test)
```

```python
pred
```

```python
plt.plot(X_test,pred)
```

```python
plt.plot(X_test,y_test)
```

```python
r2=r2_score(y_test,pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for a "U" shaped curve
#np.random.seed(42)

# Create X values
X = np.linspace(-5, 5, 100)

# Create a "U" shaped curve using a quadratic equation
y = 0.5 * X**2 + np.random.normal(scale=3, size=len(X))

# Plot the "U" shaped curve
plt.scatter(X, y, label='Synthetic Data')
plt.title('U Shaped Curve Example')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
